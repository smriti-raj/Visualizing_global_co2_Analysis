```python
# Top 10 CO₂ Emissions by CountryTop Countries
top10 = (
    df.groupby("country")["co2"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
top10.plot(kind="bar")

plt.title("Top 10 Countries by CO₂ Emissions")
plt.xlabel("Country")
plt.ylabel("CO₂")
plt.xticks(rotation=45)

plt.show()
```

```python
# CO₂ Trend Over Time
yearly = df.groupby("year")["co2"].sum()

plt.figure(figsize=(12,5))

plt.plot(yearly.index, yearly.values)

plt.title("Global CO₂ Emissions Over Time")
plt.xlabel("Year")
plt.ylabel("CO₂")

plt.grid(True)

plt.show()
```

```python
# GDP vs CO₂
plt.figure(figsize=(8,6))

plt.scatter(df["gdp"], df["co2"], alpha=0.5)

plt.xlabel("GDP")
plt.ylabel("CO₂")

plt.show()
```

```python
# Population vs CO₂
population = df.groupby("year")["population"].sum()

plt.figure(figsize=(10,5))

plt.plot(population.index, population.values, linewidth=2)

plt.title("Global Population Trend")
plt.xlabel("Year")
plt.ylabel("Population")

plt.grid(True)

plt.show()
```

```python
# CO₂ Distribution
plt.figure(figsize=(8,5))

plt.hist(df["co2"].dropna(), bins=40)

plt.title("Distribution of CO₂")

plt.xlabel("CO₂")

plt.ylabel("Frequency")

plt.show()
```

```python
## Population Distribution
plt.figure(figsize=(8,5))

plt.hist(df["population"].dropna(), bins=30)

plt.title("Population Distribution")

plt.show()
```

```python
# GDP Distribution
plt.figure(figsize=(8,5))

plt.hist(df["gdp"].dropna(), bins=30)

plt.title("GDP Distribution")

plt.show()
```

```python
# CO₂ Boxplot
plt.style.use("dark_background")

plt.figure(figsize=(6,5))

plt.boxplot(
    df["co2"].dropna(),
    patch_artist=True,
    boxprops=dict(facecolor="cyan", color="white"),
    medianprops=dict(color="yellow", linewidth=2),
    whiskerprops=dict(color="white"),
    capprops=dict(color="white")
)

plt.title("CO₂ Box Plot")

plt.show()
```

```python
## GDP vs CO₂
yearly = df.groupby("year")["co2"].sum()

plt.figure(figsize=(10,5))

plt.fill_between(
    yearly.index,
    yearly.values,
    color="purple",
    alpha=0.5
)

plt.plot(yearly.index, yearly.values, color="black")

plt.title("Global CO₂ Emissions Trend")
plt.xlabel("Year")
plt.ylabel("CO₂")

plt.show()
```

```python
## Population vs CO₂
plt.style.use("dark_background")

yearly = df.groupby("year")["co2"].sum()

plt.figure(figsize=(12,6))

plt.plot(
    yearly.index,
    yearly.values,
    color="cyan",
    linewidth=3,
    marker="o",
    markersize=5
)

plt.title("Global CO₂ Emissions", fontsize=18)
plt.xlabel("Year")
plt.ylabel("CO₂ Emissions")

plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df["gdp"], df["co2_per_capita"], alpha=0.5)

plt.xlabel("GDP")

plt.ylabel("CO₂ Per Capita")

plt.show() 
```

```python
# Global CO₂ Trend

yearly = df.groupby("year")["co2"].sum()

plt.figure(figsize=(12,5))

plt.plot(yearly.index, yearly.values)

plt.title("Global CO₂ Emissions")

plt.xlabel("Year")

plt.ylabel("CO₂")

plt.grid()

plt.show()
```

```python
## Global Population Trend

population = df.groupby("year")["population"].sum()

plt.figure(figsize=(12,5))

plt.plot(population.index, population.values)

plt.title("Global Population")

plt.show()
```

```python
# Top 10 CO₂ Countries

top10 = df.groupby("country")["co2"].sum().nlargest(10)

plt.figure(figsize=(10,6))

top10.plot(kind="bar")

plt.title("Top 10 CO₂ Emitting Countries")

plt.show()
```

```python
# Top 10 GDP Countries
top10 = df.groupby("country")["gdp"].max().nlargest(10)

plt.figure(figsize=(10,6))

top10.plot(kind="bar")

plt.title("Top 10 GDP Countries")

plt.show()
```

```python
## Top 10 Population Countries

top10 = df.groupby("country")["population"].max().nlargest(10)

plt.figure(figsize=(10,6))

top10.plot(kind="bar")

plt.title("Top 10 Population Countries")

plt.show()
```

```python
# Correlation Matrix

corr = df.select_dtypes(include="number").corr()

plt.figure(figsize=(18,15))

plt.imshow(corr)

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)

plt.yticks(range(len(corr.columns)), corr.columns)

plt.show()
```
